### Problem 1 - Numerical Differentiation

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return(np.exp(x)*np.sin(x))

def df(x): #analytical solution to check
    return(np.exp(x)*np.sin(x) + np.exp(x)*np.cos(x))
```

```python
def fd_der(f,x,h):
    return((f(x+h)-f(x))/h)

def bd_der(f,x,h):
    return((f(x)-f(x-h))/h)

def cd_der(f,x,h):
    return((f(x+h/2)-f(x-h/2))/h)
```

```python
x = 0.1

h = np.logspace(-15,-2,100)
fd =[]
bd =[]
cd =[]
fd_error = []
bd_error = []
cd_error = []

for i in range(len(h)):
    fd.append(fd_der(f,x,h[i]))
    bd.append(bd_der(f,x,h[i]))
    cd.append(cd_der(f,x,h[i]))
    fd_error.append(fd[i] - df(x))
    bd_error.append(bd[i] - df(x))
    cd_error.append(cd[i] - df(x))
    
plt.plot(h,fd,label='Forward')
plt.plot(h,bd,label='Backwards')
plt.plot(h,cd,label='Central')

plt.xlabel('h')
plt.ylabel('df(x)/dx')
plt.xscale('log')
plt.legend()
plt.title('Numerical Derivative of f(x) at x=0.1')
```

```python
plt.plot(h,np.abs(fd_error),label='Forward error')
plt.plot(h,np.abs(bd_error),label='Backwards error')
plt.plot(h,np.abs(cd_error),label='Central error')
plt.xscale('log')
plt.yscale('log')
plt.title('Error in Derivative at x=0.1')
plt.legend()
```

So round-off error starts to dominate around h = 1e-9 for forwards/backwards error, and around 1e-5 for central.

```python
import cmath
```

```python
def f_z(z):
    return(cmath.exp(2*z))

def derive_z(f,m,z,N):
    C = math.factorial(m)/N
    sum_val = 0
    k = 0
    while k < N:
        zk = cmath.exp(1j * 2* np.pi * k / N)
        new = f(zk)*cmath.exp(-1j *2*np.pi*k*m/N)
        sum_val+= new
        k+=1
    return(sum_val*C)
```

```python
for i in range(21):
    der = derive_z(f_z,i,0,10000)
    print(i,"th derivative: ",der)
```

```python
w_xy = np.asfarray(np.loadtxt('altitude.txt'))
print(w_xy)
np.shape(w_xy)
```

```python
np.shape(w_xy)

h=30000
dw_dx = []
dw_dy = []
for i in range(np.shape(w_xy)[0]):
    #print("i= ",i)
    dx =[]
    dy =[]
    if i == 0: #I deal with edge cases by wrapping around
        iplus = 1
        iminus = np.shape(w_xy)[0]-1
    elif i == np.shape(w_xy)[0]-1:
        iplus = 0
        iminus = i-1
    else:
        iplus = i+1
        iminus = i-1
    for j in range(np.shape(w_xy)[1]):
        #print('j= ',j)
        dx.append((w_xy[iplus,j] - w_xy[iminus,j])/(2*h))
        if j == 0:
            jplus = 1
            jminus = np.shape(w_xy)[1]-1
            dy.append((w_xy[i,jplus] - w_xy[i,jminus])/(2*h))
        elif j == np.shape(w_xy)[1]-1:
            jplus = 0
            jminus = j-1
            dy.append((w_xy[i,jplus] - w_xy[i,jminus])/(2*h))
        else:
            dy.append((w_xy[i,j+1] - w_xy[i,j-1])/(2*h))
    dw_dx.append(dx)
    dw_dy.append(dy)
```

```python
i = np.arange(1,10)
E = [25*(j-1) for j in i]
f = [10.6,16.0,45.0,83.5,52.8,19.9,10.8,8.25,4.7]

plt.plot(E,f,'.')
plt.xlabel('E [MeV]')
plt.ylabel('f(E) [mb]')
```

```python
def lagrange(x,X,i):
    lamb = 1
    for j in range(len(X)):
        if i!=j:
            lamb*=(x-X[j])/(X[i]-X[j])
    return(lamb)

def poly_interp(x,X,Y):
    P = []
    for i in range(len(X)):
        lamba_i = lagrange(x,X,i)
        P.append(lamba_i*Y[i])
    return(np.sum(P))
```

```python
poly_interp(120,E,f)
```

```python
x = np.linspace(0,200,400)
y = [poly_interp(i,E,f) for i in x]

plt.plot(x,y,label='Polynomial Interpolation')
plt.plot(E,f,'.')
plt.xlabel('E [MeV]')
plt.ylabel('f(E) [mb]')
plt.title('')
plt.legend()
```

```python
x = np.linspace(70,80,)
y = [poly_interp(i,E,f) for i in x]

plt.plot(x,y,label='Polynomial Interpolation')
plt.xlabel('E [MeV]')
plt.ylabel('f(E) [mb]')
plt.title('')
plt.legend()
```

So the resonance energy is at E_r = 75 +- 2 MeV. This is (78-75)/2 ~ 1.5 standard deviations away from the theoretical value 78 MeV.

From our numerical value for E_r we can find the FWHM:

```python
val = poly_interp(75,E,f) /2
print(val)
```

Reading off the plot, we see that the FWHM is then:

FWHM+ ~ 35 +- 5 MeV


FWHM- ~ 27 +- 2MeV

So:

FWHM = (35+27)/2 = 31 MeV

with error = sqrt(5^2 + 2^2) ~ 5 MeV

So our numerical value of the FWHM is (55-31)/5 = 4.8 standard deviations away from the theoretical value.
